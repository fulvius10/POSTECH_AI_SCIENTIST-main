# Lakehouse de exemplo com arquitetura Medallion

A arquitetura Medallion organiza os dados em camadas progressivas de qualidade:

No Databricks, cada camada será criada como um **schema** do Lakehouse:

- `bronze`
- `silver`
- `gold`

# Tipos de arquivos em Data Lake e Lakehouse

Em projetos de dados, o formato do arquivo influencia custo, performance, facilidade de leitura e governança. Antes de gravar Bronze, Silver e Gold, vale entender os formatos mais comuns.

## Formatos linha a linha

### CSV
CSV é um formato texto, simples e muito usado para troca de dados.

Pontos positivos:

- fácil de abrir em planilhas e editores de texto;
- bom para integração simples com sistemas legados;
- bastante conhecido.

Pontos de atenção:

- não guarda schema de forma rica;
- geralmente ocupa mais espaço;
- leitura pode ser mais lenta em grandes volumes;
- tipos como data, número e booleano precisam ser inferidos ou definidos.

Uso comum: chegada de arquivos simples na camada Bronze.

### JSON
JSON é um formato texto muito usado por APIs, eventos, logs e bancos NoSQL.

Pontos positivos:

- suporta objetos aninhados e arrays;
- combina bem com eventos semiestruturados, como DynamoDB Streams;
- é fácil de inspecionar manualmente.

Pontos de atenção:

- pode ocupar bastante espaço;
- parsing é mais custoso que formatos colunares;
- campos muito aninhados exigem tratamento com `select`, `explode`, `from_json` e acesso a structs.

Uso comum: eventos, logs, APIs, CDC semiestruturado e camada Bronze.

### Avro
Avro é um formato binário orientado a linhas, com schema embutido.

Pontos positivos:

- bom para streaming e troca entre sistemas;
- suporta evolução de schema;
- comum em pipelines com Kafka e CDC.

Pontos de atenção:

- não é tão eficiente quanto Parquet para consultas analíticas colunares;
- nem sempre é o formato final ideal para Silver e Gold.

Uso comum: ingestão, streaming, CDC e Bronze.

## Formatos colunares

### Parquet
Parquet é um formato colunar muito usado em Data Lakes.

Pontos positivos:

- excelente para consultas analíticas;
- lê apenas as colunas necessárias;
- comprime bem;
- guarda schema;
- funciona muito bem com Spark.

Pontos de atenção:

- não traz sozinho controle transacional ACID;
- updates, deletes e merges exigem mecanismos adicionais ou regravações.

Uso comum: Silver e Gold quando não há necessidade de transação Delta.

### ORC
ORC também é um formato colunar. Ele é comum no ecossistema Hive.

Pontos positivos:

- boa compressão;
- eficiente para consultas analíticas;
- guarda estatísticas úteis para otimização.

Pontos de atenção:

- em Databricks e Spark moderno, Parquet e Delta costumam aparecer com mais frequência;
- a escolha geralmente depende do padrão da empresa.

Observação: quando alguém escreve “PRC”, normalmente está se referindo a **ORC** ou a **Parquet**. No Spark, os formatos comuns são `parquet` e `orc`.

## Formato de tabela Lakehouse

### Delta Lake
Delta Lake não é apenas um arquivo. Ele é uma camada de tabela sobre arquivos Parquet, com um log transacional chamado `_delta_log`.

Pontos positivos:

- transações ACID;
- suporte a `UPDATE`, `DELETE` e `MERGE`;
- versionamento e histórico;
- leitura eficiente com Parquet por baixo;
- ótimo encaixe com Lakehouse e arquitetura Medallion.

Pontos de atenção:

- é mais do que um arquivo único: é uma pasta de tabela com arquivos Parquet e log Delta;
- deve ser acessado como tabela Delta, não como um Parquet qualquer.

Uso comum: Bronze, Silver e Gold no Databricks.

## Como escolher no tutorial

Neste notebook, vamos usar:

- **JSON** para simular origens transacionais e semiestruturadas;
- **Delta** para gravar as tabelas do Lakehouse nas camadas Bronze, Silver e Gold;
- exemplos didáticos de **CSV**, **Parquet**, **ORC** e **Delta** para comparar leitura e escrita.

Regra prática para explicar em aula:

- se o dado chegou bruto e semiestruturado, JSON pode aparecer na Bronze;
- se a consulta é analítica e colunar, Parquet é eficiente;
- se precisa de tabela confiável, histórico e operações como merge, use Delta;
- se a empresa usa Hive/ORC como padrão, ORC pode aparecer em integrações específicas.

# Origem dos dados — DynamoDB, CDC/DMS e arquivos

Antes de criar as camadas Medallion, vamos simular três formas comuns de chegada de dados em um Lakehouse:

## 1. Sistema transacional estilo DynamoDB
O DynamoDB é um banco NoSQL usado em aplicações transacionais. Os dados costumam chegar ao Lakehouse por eventos, por exportações ou por streams.

Neste tutorial, vamos simular eventos no formato parecido com **DynamoDB Streams**, com campos aninhados como `NewImage`, atributos tipados (`S`, `N`, `NULL`) e uma lista de itens da venda.

Para transformar esse JSON em tabela, vamos usar:

- `spark.createDataFrame()` para criar uma coluna com o JSON bruto sem usar `sparkContext`;
- `from_json()` para interpretar o texto JSON usando um schema;
- `explode()` para abrir a lista de itens da venda;
- acesso a campos aninhados como `dynamodb.NewImage.id_venda.N`;
- `cast()` para converter texto em número.

## Observação importante sobre atributos tipados do DynamoDB

Nos eventos do DynamoDB, os valores costumam vir acompanhados do tipo do atributo. Por isso aparecem marcações como `S`, `N` e `NULL`.

Exemplo simplificado:

```json
{
  "id_venda": { "N": "1" },
  "produto": { "S": "Notebook" },
  "cupom": { "NULL": true }
}
```

Significado:

- `S` significa **String**, ou seja, texto. Exemplo: `{ "S": "Notebook" }`.
- `N` significa **Number**, ou seja, número. No JSON do DynamoDB, o número vem como texto, por exemplo `{ "N": "1" }`, então no Spark usamos `cast()` para converter.
- `NULL` indica valor nulo, por exemplo `{ "NULL": true }`.

Por isso, para pegar o `id_venda`, usamos algo como:

```python
F.col("dynamodb.NewImage.id_venda.N").cast("int")
```

A leitura é: entrar em `id_venda`, pegar o valor dentro de `N` e converter para inteiro.

# Camada Bronze — dados brutos no Lakehouse

A camada Bronze recebe os dados como chegaram ao pipeline, preservando rastreabilidade e histórico de ingestão.

Neste exemplo, vamos gravar quatro tipos de tabela Bronze:

- eventos brutos simulados de DynamoDB Streams;
- eventos brutos simulados de CDC/DMS;
- vendas já explodidas a partir do JSON transacional;
- clientes consolidados a partir dos eventos CDC/DMS.

Na prática, é comum manter tanto o bruto original quanto uma primeira versão estruturada para facilitar reprocessamentos e auditoria.


````python
# Camada Bronze: cria schemas e grava dados brutos no Lakehouse

catalogo_atual = spark.sql("SELECT current_catalog() AS catalogo").first()["catalogo"]
schema_atual = spark.sql("SELECT current_schema() AS schema").first()["schema"]

print("Catálogo atual:", catalogo_atual)
print("Schema atual:", schema_atual)

for camada in ["bronze", "silver", "gold"]:
    spark.sql(f"CREATE SCHEMA IF NOT EXISTS {camada}")

bronze_dynamodb_vendas_raw = "bronze.aula3_dynamodb_vendas_raw"
bronze_dms_clientes_raw = "bronze.aula3_dms_clientes_raw"
bronze_vendas = "bronze.aula3_vendas_raw"
bronze_clientes = "bronze.aula3_clientes_raw"

(
    df_raw_dynamodb_vendas
    .withColumn("_data_ingestao", F.current_timestamp())
    .withColumn("_fonte", F.lit("dynamodb_stream_simulado"))
    .write
    .format("delta")
    .mode("overwrite")
    .option("overwriteSchema", "true")
    .saveAsTable(bronze_dynamodb_vendas_raw)
)

(
    df_raw_dms_clientes
    .withColumn("_data_ingestao", F.current_timestamp())
    .withColumn("_fonte", F.lit("dms_cdc_simulado"))
    .write
    .format("delta")
    .mode("overwrite")
    .option("overwriteSchema", "true")
    .saveAsTable(bronze_dms_clientes_raw)
)

df_bronze_vendas = (
    df_vendas
    .withColumn("_data_ingestao", F.current_timestamp())
    .withColumn("_fonte", F.lit("dynamodb_stream_explodido"))
)

df_bronze_clientes = (
    df_clientes
    .withColumn("_data_ingestao", F.current_timestamp())
    .withColumn("_fonte", F.lit("dms_cdc_consolidado"))
)

(
    df_bronze_vendas
    .write
    .format("delta")
    .mode("overwrite")
    .option("overwriteSchema", "true")
    .saveAsTable(bronze_vendas)
)

(
    df_bronze_clientes
    .write
    .format("delta")
    .mode("overwrite")
    .option("overwriteSchema", "true")
    .saveAsTable(bronze_clientes)
)

print("Tabelas Bronze criadas:")
for tabela in [bronze_dynamodb_vendas_raw, bronze_dms_clientes_raw, bronze_vendas, bronze_clientes]:
    print("-", tabela)

display(spark.read.table(bronze_vendas))
````

````python
# unionByName() combina bases pelo nome das colunas
# A base abaixo simula novas vendas com a mesma estrutura, mas em ordem diferente
dados_vendas_extra = [
    ("Loja-PR", None, 109, "PR", 500.0, "Moveis", 1, "Mesa Lateral", 12, "2026-01-08", "ok"),
    ("Loja-SP", "CUPOM15", 101, "SP", 220.0, "Eletronicos", 1, "Mouse Gamer", 13, "2026-01-08", "ok")
]

colunas_extra = [
    "origem", "cupom", "id_cliente", "estado", "preco_unitario",
    "categoria", "quantidade", "produto", "id_venda", "data_venda", "status_bruto"
]

df_vendas_extra = spark.createDataFrame(dados_vendas_extra, schema=colunas_extra)

df_extra_tratado = (
    df_vendas_extra
    .withColumnRenamed("data_venda", "dt_venda")
    .withColumn("valor_total", F.col("quantidade") * F.col("preco_unitario"))
    .withColumn("canal", F.lit("Loja Online"))
    .withColumn(
        "faixa_valor",
        F.when(F.col("valor_total") >= 3000, F.lit("Alta"))
         .when(F.col("valor_total") >= 1000, F.lit("Media"))
         .otherwise(F.lit("Baixa"))
    )
    .withColumn("status_tratado", F.trim(F.regexp_replace(F.col("status_bruto"), "-", " ")))
    .withColumn("dt_venda", F.to_date(F.col("dt_venda"), "yyyy-MM-dd"))
    .withColumn("estado_final", F.coalesce(F.col("estado"), F.lit("NAO_INFORMADO")))
    .withColumn("origem_padronizada", F.regexp_replace(F.col("origem"), "-", "_"))
    .fillna({"preco_unitario": 0.0, "cupom": "SEM_CUPOM"})
    .drop("estado", "status_bruto")
)

df_unificado = df_sem_duplicidade.unionByName(df_extra_tratado, allowMissingColumns=True)
display(df_unificado.orderBy("id_venda"))
````

````python
# groupBy() + agg() gera métricas agregadas
df_analitico_categoria = (
    df_enriquecido
    .groupBy("categoria")
    .agg(
        F.count("*").alias("qtd_registros"),
        F.sum("valor_total").alias("faturamento_total"),
        F.avg("valor_total").alias("ticket_medio"),
        F.countDistinct("id_cliente").alias("clientes_distintos")
    )
    .withColumn("faturamento_total", F.round(F.col("faturamento_total"), 2))
    .withColumn("ticket_medio", F.round(F.col("ticket_medio"), 2))
    .orderBy(F.desc("faturamento_total"))
)

display(df_analitico_categoria)
````

````python
# Outro corte analítico por estado e segmento
df_analitico_estado_segmento = (
    df_enriquecido
    .groupBy("estado_final", "segmento")
    .agg(
        F.sum("valor_total").alias("faturamento_total"),
        F.count("*").alias("qtd_vendas")
    )
    .withColumn("faturamento_total", F.round(F.col("faturamento_total"), 2))
    .orderBy("estado_final", F.desc("faturamento_total"))
)

display(df_analitico_estado_segmento)
````

# Configurações e boas práticas para reduzir shuffle e skew

Antes de gravar ou analisar performance, é importante entender dois problemas comuns em Spark:

## Shuffle
Shuffle acontece quando o Spark precisa redistribuir dados entre partições. Ele aparece com frequência em operações como `join()`, `groupBy()`, `distinct()`, `dropDuplicates()`, `orderBy()`, `repartition()` e funções de janela.

Nem todo shuffle é ruim: muitas análises precisam dele. O objetivo é **evitar shuffle desnecessário** e tornar os shuffles inevitáveis mais eficientes.

## Skew
Skew acontece quando uma ou poucas chaves concentram muitos registros. Exemplo: muitas vendas no estado `SP` e poucas nos demais estados. Isso cria partições muito maiores que outras, fazendo algumas tarefas demorarem muito mais.

## Configurações importantes

### `spark.sql.adaptive.enabled`
Ativa o **Adaptive Query Execution (AQE)**. Com AQE, o Spark pode ajustar o plano durante a execução, depois de observar estatísticas reais dos dados.

### `spark.sql.adaptive.coalescePartitions.enabled`
Permite que o AQE junte partições pequenas após shuffle. Ajuda a reduzir overhead quando existem partições demais para pouco dado.

### `spark.sql.adaptive.skewJoin.enabled`
Permite que o AQE tente tratar joins com skew, dividindo partições muito grandes em partes menores.

### `spark.sql.shuffle.partitions`
Define o número padrão de partições após shuffle. O padrão costuma ser alto para bases pequenas de aula. Em produção, esse número deve ser escolhido conforme volume, tamanho dos arquivos, quantidade de workers e tempo de execução.

### `spark.sql.autoBroadcastJoinThreshold`
Define até que tamanho uma tabela pode ser enviada por broadcast automaticamente. Broadcast é útil quando uma tabela pequena será usada em join com uma tabela maior, evitando redistribuir os dois lados do join.

## Pontos importantes para citar em aula

- Filtre linhas e selecione colunas antes de `join()` e `groupBy()`.
- Use `broadcast()` quando uma tabela pequena será combinada com uma tabela maior.
- Evite `orderBy()` global sem necessidade, porque ele costuma gerar shuffle caro.
- Prefira `coalesce()` para reduzir partições antes de salvar poucos arquivos; use `repartition()` quando precisar redistribuir os dados.
- Ao fazer `join()` ou agregações, observe se a chave é muito concentrada em poucos valores.
- Para skew severo, considere técnicas como salting, pré-agregação, broadcast da dimensão pequena ou mudança da chave de particionamento.
- Use `explain("formatted")` para verificar se o Spark está usando `BroadcastHashJoin`, `SortMergeJoin`, `Exchange`, `AdaptiveSparkPlan` e outras etapas do plano.
- Não existe configuração mágica: ajuste com base no volume, no plano de execução e no tempo observado.


````python
# Configurações úteis para a aula
# Em Databricks Free Edition, o ambiente é serverless, mas estas configs SQL podem ser ajustadas na sessão do notebook.

spark.conf.set("spark.sql.adaptive.enabled", "true")
spark.conf.set("spark.sql.adaptive.coalescePartitions.enabled", "true")
spark.conf.set("spark.sql.adaptive.skewJoin.enabled", "true")

# Para uma base pequena de aula, poucas partições de shuffle são suficientes.
# Em produção, não use este número no automático: avalie volume de dados, workers e tamanho das partições.
spark.conf.set("spark.sql.shuffle.partitions", "8")

# Permite broadcast automático de tabelas pequenas até 50 MB.
# Aumentar demais este valor pode causar pressão de memória.
spark.conf.set("spark.sql.autoBroadcastJoinThreshold", str(50 * 1024 * 1024))

configs_importantes = [
    "spark.sql.adaptive.enabled",
    "spark.sql.adaptive.coalescePartitions.enabled",
    "spark.sql.adaptive.skewJoin.enabled",
    "spark.sql.shuffle.partitions",
    "spark.sql.autoBroadcastJoinThreshold"
]

for chave in configs_importantes:
    print(chave, "=", spark.conf.get(chave))

# Exemplo 1: broadcast explícito da tabela pequena de clientes
# Neste tutorial, df_clientes_tratado é pequeno, então faz sentido enviá-lo para todos os executores.
df_enriquecido_broadcast = df_unificado.join(
    F.broadcast(df_clientes_tratado),
    on="id_cliente",
    how="left"
)

display(df_enriquecido_broadcast)

# Exemplo 2: reparticionar por uma chave usada em agregação ou join
# Use com cuidado: repartition() também gera shuffle, então ele só compensa quando melhora etapas seguintes.
df_particionado_estado = df_enriquecido.repartition("estado_final")
print("Partições após repartition por estado_final:", df_particionado_estado.rdd.getNumPartitions())

# Exemplo 3: validar o plano de execução
# Procure termos como AdaptiveSparkPlan, Exchange, BroadcastHashJoin e SortMergeJoin.
df_enriquecido_broadcast.explain("formatted")
````

## Estatística rápida e performance
- `sample()` cria amostra aleatória
- `approxQuantile()` calcula quantis aproximados
- `cache()` mantém dados em memória para reuso
- `repartition()` redistribui partições
- `coalesce()` reduz partições
- `rdd.getNumPartitions()` mostra quantas partições existem
- `explain("formatted")` e `explain("extended")` mostram o plano de execução

- `spark.sql.adaptive.skewJoin.enabled` ajuda no tratamento de joins com skew
- `spark.sql.shuffle.partitions` define partições padrão de shuffle
- `spark.sql.autoBroadcastJoinThreshold` controla broadcast automático
- `broadcast()` força broadcast de uma tabela pequena em joins

## Lakehouse e arquitetura Medallion
- `CREATE SCHEMA IF NOT EXISTS` cria as camadas Bronze, Silver e Gold

- `spark.read.table()` lê tabelas do Lakehouse
- Bronze mantém dados brutos e metadados de ingestão

## Origens simuladas
- DynamoDB Streams foi simulado com JSON aninhado e atributos tipados

- `explode()` abre arrays de objetos, como a lista de itens de uma venda
- CDC/DMS foi simulado com operações `I`, `U` e timestamp de commit
- `row_number()` ajuda a consolidar o estado mais recente de uma entidade no CDC
- Arquivos JSON foram simulados como JSON Lines em memória para evitar dependência de `dbfs:/tmp`

## Tipos de arquivos
- CSV é simples, textual e comum em integrações, mas não preserva schema rico
- JSON é útil para eventos, APIs, logs e objetos aninhados
- Avro é binário, orientado a linhas e comum em streaming/CDC
- Parquet é colunar, comprimido e eficiente para consultas analíticas
- ORC é colunar e comum em ambientes Hive
- Delta Lake usa Parquet com log transacional `_delta_log`
- `write.format(...).save()` grava arquivos em uma pasta
- `write.format("delta").saveAsTable()` grava tabelas Delta gerenciadas no Lakehouse
