# Aprendendo probabilidade na prática!


Em ciência de dados, compreender o comportamento dos dados é essencial para gerar análises precisas e previsões confiáveis. No entanto, o mundo real é repleto de incertezas: nem sempre conseguimos prever o comportamento exato de um cliente, o número de compras em um dia ou o tempo que uma entrega levará. É nesse cenário que entram a probabilidade e as distribuições de probabilidade, que são ferramentas fundamentais que nos permitem quantificar a incerteza e modelar matematicamente o acaso.

A probabilidade **descreve a chance de um evento ocorrer**, enquanto as **distribuições mostram como essas probabilidades se espalha por todos os valores possíveis de uma variável aleatória**. Com esse conhecimento, o(a) cientista de dados é capaz de entender padrões, realizar inferências e apoiar decisões baseadas em evidências. A probabilidade mede a chance de um evento ocorrer, variando de 0 a 1:

## **Propriedades da probabilidade**

### **Não existe probabilidade negativa ⚠️**

A probabilidade de qualquer evento **sempre será maior que 0**, ou seja, sempre positiva.

### **As somas de todas as probabilidades é igual a 1 🎲**

Toda vez que estamos analisando e calculando probabilidades, sempre a soma deve ser igual a 1. Vamos a um exemplo prático! Um dado comum tem 6 faces (1 a 6). Cada face tem a mesma chance de sair, então:

P(1)+P(2)+P(3)+P(4)+P(5)+P(6)



image.png

### **Evento impossível** ❌

A probabilidade de um evento impossível é igual a 0. Por exemplo, ao lançar uma moeda, só existem dois resultados possíveis: cara ou coroa. Portanto, qualquer outro evento, como sair “número 3”, tem probabilidade zero, pois não pode acontecer dentro desse experimento.

### **Evento certo ✅**

Um evento certo é aquele que sempre acontece, ou seja, sua probabilidade é igual a 1. Em outras palavras, é um evento cuja ocorrência é garantida, não há incerteza envolvida. Imagine um modelo que analisa o comportamento de clientes de um e-commerce.  Um evento certo poderia ser: “Todo cliente que fez uma compra tem um registro na base de pedidos.” Isso é garantido pelo próprio funcionamento do sistema.

### **Adição de probabilidade (eventos mutuamente exclusivos)** ➕

A adição de probabilidade ocorre quando dois eventos são mutuamente exclusivos (ou disjuntos) e não podem acontecer ao mesmo tempo. Isso significa que a interseção entre eles é nula:

image.png

Vamos para um exemplo: Em uma base de clientes de um banco:

A: o cliente foi aprovado no pedido de cartão: P(A)=0.7

B: o cliente foi reprovado: P(B)=0.3

Um cliente não pode ser aprovado e reprovado ao mesmo tempo, logo:





image.png

A soma das probabilidades cobre todos os resultados possíveis, o cliente será aprovado ou reprovado, sem sobreposição.

### **Multiplicação de probabilidade (eventos independentes)** 🧮

Se dois eventos ocorrem simultaneamente (o resultado de um não afeta o outro). Nesses casos, usamos a regra da multiplicação para eventos independentes:Essa lógica é aplicada, por exemplo, quando queremos calcular a probabilidade conjunta de dois atributos independentes ocorrerem em um modelo probabilístico (como o Naive Bayes).

Por exemplo: a probabilidade de um cliente abrir um e-mail promocional (A) e comprar um produto (B), assumindo que as ações são independentes:

* P(A)=0.4  (40% dos clientes abrem o e-mail)
* P(B)=0.2  (20% dos clientes compram o produto)

Logo, a probabilidade de um cliente abrir o e-mail e comprar o produto é de 8%:



image.png

image.png

### **Probabilidade complementar** ➖

A probabilidade complementar é usada quando queremos saber a chance de um evento não acontecer. Então calculamos que o evento A, por exemplo, é igual a 1 menos a probabilidade de A acontecer. Suponha que a probabilidade de um cliente comprar um produto após abrir o e-mail seja P(A)=0,3. Então temos 70% de chance de o cliente não comprar o produto. É como pensar que tudo pode acontecer (representado por 1), menos o que não pode acontecer (nesse caso representado por 0,3).

# **Probabilidade condicional**

A probabilidade condicional é um dos fundamentos teóricos mais utilizados em modelos probabilísticos de classificação, análise de comportamento de usuários, detecção de fraudes, e em diversos outros domínios da ciência de dados. Sua importância decorre do fato de que grande parte dos fenômenos estudados nessa área envolve eventos dependentes, em que o conhecimento prévio de uma variável altera a probabilidade de outra.

**Exemplo prático: comportamento de clientes** 📩

Suponha que temos dados de clientes sobre:

- se abriram o e-mail

- se compraram o produto

Quero saber qual é a probabilidade de um cliente comprar um produto, dado que ele abriu um e-mail. P(Comprar∣Abriu o e-mail):

image.png

P(B | A) → probabilidade de o evento B acontecer condicionada ao fato de que A já aconteceu

P(A ∩ B) → probabilidade de A e B acontecerem juntos

P(A) → probabilidade de A acontecer

```python
import pandas as pd

# Criando um dataset simples
data = {
    'abriu_email': [1, 1, 1, 1, 0, 0, 0, 0, 1, 0],
    'comprou':     [1, 0, 1, 0, 0, 1, 0, 0, 1, 0]
}

df = pd.DataFrame(data)

df.head()
```

```python
abriu_email = df['abriu_email'].sum()
abriu_email
```

```python
total = len(df)
total
```

```python
P_A_racional = abriu_email / total
P_A_racional
```

```python
P_A = df['abriu_email'].mean() #calculando a probabilidade de abrir o e-mail
P_A
```

```python
P_A_e_B = df[(df['abriu_email'] == 1) & (df['comprou'] == 1)].shape[0] / len(df) # Probabilidade conjunta: abrir e comprar
P_A_e_B
```

```python
P_B_dado_A = P_A_e_B / P_A # Probabilidade condicional: comprar dado que abriu
P_B_dado_A
```

Concluímos que: 60% dos clientes que abriram o e-mail acabaram comprando o produto

# Probabilidade independente

Na probabilidade independente, temos dois eventos A e B que são independentes se o fato de um ocorrer, não afetar a probabilidade do outro ocorrer. E a principal consequência disso é a regra da multiplicação:

image.png

Ou seja:

- Saber que B aconteceu não muda nada sobre A

- As probabilidades “não conversam” entre si

Exemplo prático: Vamos simular dois eventos independentes

- Abrir um e-mail promocional

- Comprar um produto independentemente de ter aberto o e-mail



```python
import numpy as np

np.random.seed(42)

n = 1000

df = pd.DataFrame({
    'abriu_email': np.random.binomial(1, 0.4, n),  # 40% abrem
    'comprou': np.random.binomial(1, 0.2, n)       # 20% compram (independente)
})

df.head()
```

```python
# Calculando as probabilidades

P_A = df['abriu_email'].mean()
P_B = df['comprou'].mean()

P_A, P_B
```

Regra da multiplicação válida para eventos independentes.

```python
P_A_vezes_B = P_A * P_B # Probabilidade conjunta teórica (regra da multiplicação)
P_A_vezes_B
```

cerca de 7,9% de chance

# DISTRIBUIÇÕES DISCRETAS

Distribuições discretas descrevem situações em que a variável aleatória só pode assumir valores contáveis, geralmente inteiros.

Ou seja:
👉 você consegue listar todos os valores possíveis, mesmo que a lista seja infinita.

📌**Distribuição uniforme:** é aquela em que **todos os resultados possíveis têm a mesma probabilidade de ocorrer**. Em outras palavras, nenhum valor é mais provável que outro. Todos os valores têm a mesma probabilidade.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats


sns.set(style="whitegrid")
```

```python
valores = np.random.randint(1, 7, size=1000)


sns.countplot(x=valores)
plt.title("Distribuição Uniforme Discreta — Dado")
plt.show()
```

O gráfico mostra uma distribuição uniforme discreta, onde as frequências das faces do dado (Cada barra representa quantas vezes aquela face saiu após vários lançamentos) são semelhantes, indicando que cada face tem aproximadamente a mesma probabilidade de ocorrer, próxima de 1/6. Como as barras têm alturas parecidas, concluímos que o dado é justo.

Probabilidade ≈ frequência relativa (todas próximas de 0,1667%)

Quando usar?

- Sorteios

- Simulações simples

- Experimentos sem viés

📌 **Distribuição Binomial:** Trata-se de uma distribuição discreta usada para representar o número de sucessos obtidos em um determinado número de tentativas. Ela se caracteriza por apresentar **apenas duas possíveis saídas: sucesso ou fracasso**. Um exemplo clássico é a probabilidade de obter cara ou coroa ao lançar uma moeda.

*Parâmetros:*

*n: número de tentativas*

*p: probabilidade de sucesso*

Cenário de lançar uma moeda (cara ou coroa)

Aqui estamos assumindo algo como:

- Lançar uma moeda justa

- n = 10 lançamentos

- “Sucesso” = sair cara

Probabilidade de cara em cada lançamento:

𝑝 = 0,5

A variável aleatória é X (número de caras em 10 lançamentos)

O eixo Y é a probabilidade de acontecer exatamente aquele número de caras.

```python
n = 10  # número de tentativas
p = 0.5 # probabilidade de sucesso
x = np.arange(0, n+1)


prob = stats.binom.pmf(x, n, p)


plt.bar(x, prob)
plt.title("Distribuição Binomial")
plt.xlabel("Número de sucessos")
plt.ylabel("Probabilidade")
plt.show()
```

A distribuição binomial mostra a probabilidade de obter exatamente
𝑘 caras ao lançar uma moeda 10 vezes, sendo os valores centrais mais prováveis e os extremos menos frequentes.

Quando usar?

- Conversão

- A/B tests

- Taxa de resposta

📌 **Distribuição de poisson:** A Distribuição de Poisson é um tipo de distribuição de probabilidade discreta usada para **modelar o número de vezes que um evento ocorre em um intervalo fixo de tempo**, espaço ou área, quando esses eventos acontecem de forma independente e com uma taxa média constante. Em outras palavras, podemos dizer que esse tipo de distribuição responde algo como: “Quantas vezes algo vai acontecer em um determinado período, se em média isso ocorre λ vezes?”

*Parâmetro:*

*λ (lambda): taxa média de ocorrência*

Exemplo: chamadas por hora em um call center.

```python
lambda_ = 4
x = np.arange(0, 15)


prob = stats.poisson.pmf(x, lambda_)


plt.bar(x, prob)
plt.title("Distribuição de Poisson")
plt.xlabel("Número de eventos")
plt.ylabel("Probabilidade")
plt.show()
```

Estamos contando quantas chamadas chegam em 1 hora

Os eventos são:

- independentes

- podem acontecer a qualquer momento

- Existe uma taxa média de chamadas. Essa taxa é o λ (lambda).

Pelo formato do gráfico, parece algo como:

𝜆 ≈ 4 chamadas por hora

A barra mais alta está em 3 e 4 chamadas, são os cenários mais prováveis.


Esse gráfico responde perguntas como:

“Quantas chamadas devo esperar na maioria das horas?”

“Qual a chance de termos mais de 7 chamadas numa hora?”

“Preciso reforçar atendimento para picos?”

Quando usar:

- Incidentes

- Reclamações

- Acessos ao sistema

# DISTRIBUIÇÕES CONTÍNUAS

Distribuições contínuas modelam variáveis que podem assumir qualquer valor dentro de um intervalo.

Não dá pra contar valor por valor. Sempre existe algo entre dois números.

👉 Diferente das discretas (0, 1, 2…), aqui temos:

1,2

1,23

1,2345

… infinitos valores possíveis

📌 **Distribuição normal:** A distribuição normal, também chamada de distribuição gaussiana, é uma distribuição de probabilidade contínua muito usada em estatística e ciência de dados. É a famosa curva que forma um sino!  **A distribuição normal descreve como os dados se distribuem em torno da média**. A maioria dos valores tende a se concentrar próxima da média, e a probabilidade vai diminuindo à medida que os valores se afastam dela, formando a clássica curva em formato de sino.

*Parâmetros:*

*μ (média)*

*σ (desvio padrão)*

**Tipos de assimetria:**

**Assimetria positiva (ou à direita):**

- A cauda da curva se estende para a direita.

- A maioria dos valores está concentrada à esquerda da média.

- A média é maior que a mediana.

Exemplo: renda da população (poucas pessoas ganham muito mais que a maioria).


**Assimetria negativa (ou à esquerda):**

- A cauda da curva se estende para a esquerda.

- A maioria dos valores está à direita da média.

- A média é menor que a mediana.

Exemplo: idade de aposentadoria (a maioria se aposenta perto do limite, poucos muito antes).


image.png

Exemplo: notas de clientes de NPS (contínuo).

```python
dados = np.random.normal(loc=7, scale=1.5, size=1000)


sns.histplot(dados, kde=True)
plt.title("Distribuição Normal")
plt.show()
```

Por que ela é tão importante?

Teorema Central do Limite

Muitos fenômenos reais se aproximam dela

Base para vários modelos estatísticos

📌 **Distribuição Exponencial:** É uma distribuição de probabilidade contínua, **usada para representar situações em que as probabilidades diminuem de forma exponencial ao longo do tempo**. Ela é especialmente útil para modelar o tempo até que um determinado evento ocorra. Por exemplo, pode ser aplicada para estimar a probabilidade de um aparelho celular continuar funcionando após alguns anos de uso, à medida que o tempo passa, essa probabilidade não reduz de maneira constante (linear), mas sim de forma acelerada (exponencial).

*Parâmetro:*

*λ: taxa*

Exemplo: tempo até o próximo atendimento. Queremos responder "Quanto tempo eu espero até o próximo evento acontecer?"

- Evento = início do próximo atendimento

- Variável = tempo de espera

- Tempo é contínuo (segundos, minutos…)

Se 𝜆 = 1 minuto de atendimento

👉 tempo médio de espera = 1 minuto

```python
dados = np.random.exponential(scale=2, size=1000)


sns.histplot(dados, kde=True)
plt.title("Distribuição Exponencial")
plt.show()
```

A distribuição exponencial modela o tempo até o próximo atendimento, mostrando que esperas curtas são mais comuns e esperas longas são raras, mas possíveis.

Quando usar:

- Tempo de espera

- Churn

- Falhas de sistema

📌 **Distribuição t de Student:** é uma distribuição de probabilidade contínua muito usada em estatística inferencial, especialmente quando trabalhamos com **amostras pequenas (n < 30) e desconhecemos o desvio padrão da população. Ela é semelhante à distribuição normal, mas tem caudas mais largas (ou seja, mais probabilidade nas extremidades). Isso significa que a t de Student é mais “conservadora”, ela considera uma maior incerteza nos resultados, o que é importante quando a amostra é pequena**. À medida que o tamanho da amostra aumenta, a distribuição t se aproxima da distribuição normal.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t, norm

# Criando valores no eixo x
x = np.linspace(-5, 5, 1000)

# Três distribuições t com diferentes graus de liberdade
y1 = t.pdf(x, df=2)
y2 = t.pdf(x, df=10)
y3 = t.pdf(x, df=30)

# Distribuição normal para comparação
y_normal = norm.pdf(x, 0, 1)

# Plotando o gráfico
plt.plot(x, y1, label='t (df=2)', color='red')
plt.plot(x, y2, label='t (df=10)', color='green')
plt.plot(x, y3, label='t (df=30)', color='blue')
plt.plot(x, y_normal, label='Normal', color='black', linestyle='--')

plt.title("Distribuição t de Student vs Normal")
plt.xlabel("Valores")
plt.ylabel("Densidade de probabilidade")
plt.legend()
plt.grid(True)
plt.show()
```

A distribuição t de Student é usada quando:

- Queremos inferir sobre a média

- A amostra é pequena

- O desvio padrão populacional é desconhecido

Ela é tipo uma Normal com cautela 😄

O papel dos graus de liberdade (df)
𝑑𝑓 = 𝑛 − 1

- amostra pequena -> df pequeno, mais incerteza

- amostra grande -> df grande, menos incerteza

A t “se espalha” mais quando sabemos pouco sobre a população.

Quando usar:

- Média de NPS com poucas respostas

- Tempo médio de atendimento em um novo produto

- Ticket médio em uma amostra pequena

A distribuição t de Student é semelhante à normal, mas com caudas mais largas, refletindo maior incerteza quando trabalhamos com amostras pequenas.
