
# Aula Prática — Correlação, Causa e Causalidade em Ciência de Dados

## Objetivo

Nesta aula prática iremos aprender:

- Diferença entre correlação e causalidade
- Como variáveis podem parecer relacionadas sem existir causa
- Como identificar relações causais
- Como interpretar variáveis confundidoras
- Como evitar conclusões erradas em Machine Learning e Estatística

---

# Conceitos Fundamentais

## Correlação

Correlação indica que duas variáveis possuem comportamento semelhante.

Exemplo:
- Quando aumenta temperatura, aumenta venda de sorvete.

Mas isso NÃO significa obrigatoriamente que:
- Sorvete causa calor.

---

## Causalidade

Causalidade significa que uma variável realmente influencia outra.

Exemplo:
- Fumar aumenta risco de câncer.

Neste caso existe relação causal.

---

# Importante

## Correlação NÃO implica causalidade.

Essa é uma das frases mais importantes da Ciência de Dados.


```

# Bibliotecas

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

np.random.seed(42)
```


# EXEMPLO 1 — Correlação sem causalidade

## Cenário

Vamos criar um exemplo onde:

- Venda de sorvete aumenta
- Afogamentos aumentam

Aparentemente existe relação.

Mas na verdade:
- O calor é a variável escondida.


```

# Gerando temperatura

temperatura = np.random.normal(30, 5, 100)

# Sorvete depende da temperatura
venda_sorvete = temperatura * 10 + np.random.normal(0, 10, 100)

# Afogamentos também dependem da temperatura
afogamentos = temperatura * 2 + np.random.normal(0, 5, 100)

dados = pd.DataFrame({
    'Temperatura': temperatura,
    'Venda_Sorvete': venda_sorvete,
    'Afogamentos': afogamentos
})

dados.head()
```


# Visualizando os dados


```

plt.figure(figsize=(8,5))

plt.scatter(dados['Venda_Sorvete'], dados['Afogamentos'])

plt.xlabel('Venda de Sorvete')
plt.ylabel('Afogamentos')

plt.title('Correlação entre Sorvete e Afogamentos')

plt.show()
```


# Correlação Estatística


```

correlacao = dados[['Venda_Sorvete', 'Afogamentos']].corr()

print(correlacao)
```


# Interpretação

Existe forte correlação.

Mas:
- Sorvete NÃO causa afogamentos.

A variável real é:
- Temperatura

Isso é chamado de:

## Variável Confundidora



# EXEMPLO 2 — Relação Causal

## Cenário

Agora vamos criar um exemplo onde:

- Mais horas de estudo
- Melhor nota

Aqui existe uma relação causal plausível.


```

horas_estudo = np.random.normal(5, 2, 100)

nota = horas_estudo * 1.5 + np.random.normal(0, 1, 100)

dados_estudo = pd.DataFrame({
    'Horas_Estudo': horas_estudo,
    'Nota': nota
})

dados_estudo.head()
```


# Visualização


```

plt.figure(figsize=(8,5))

plt.scatter(dados_estudo['Horas_Estudo'], dados_estudo['Nota'])

plt.xlabel('Horas de Estudo')
plt.ylabel('Nota')

plt.title('Horas de Estudo vs Nota')

plt.show()
```


# Treinando modelo de regressão linear


```

X = dados_estudo[['Horas_Estudo']]
y = dados_estudo['Nota']

modelo = LinearRegression()

modelo.fit(X, y)

pred = modelo.predict(X)

r2 = r2_score(y, pred)

print(f'Coeficiente Angular: {modelo.coef_[0]:.4f}')
print(f'Intercepto: {modelo.intercept_:.4f}')
print(f'R² Score: {r2:.4f}')
```


# Linha de regressão


```

plt.figure(figsize=(8,5))

plt.scatter(X, y)

plt.plot(X, pred)

plt.xlabel('Horas de Estudo')
plt.ylabel('Nota')

plt.title('Regressão Linear')

plt.show()
```


# Interpretação

Neste cenário existe uma hipótese causal plausível:

- Mais estudo tende a gerar melhor desempenho.

Mas mesmo assim:
- Precisamos validar contexto
- Precisamos considerar outras variáveis
- Precisamos evitar conclusões precipitadas



# EXEMPLO 3 — Variável Confundidora

## Cenário

Queremos analisar:

- Exercício físico
- Perda de peso

Mas existe uma variável escondida:
- Alimentação


```

alimentacao = np.random.normal(7, 2, 100)

exercicio = alimentacao * 0.8 + np.random.normal(0, 1, 100)

perda_peso = (
    exercicio * 1.2 +
    alimentacao * 2 +
    np.random.normal(0, 2, 100)
)

dados_saude = pd.DataFrame({
    'Alimentacao': alimentacao,
    'Exercicio': exercicio,
    'Perda_Peso': perda_peso
})

dados_saude.head()
```


# Matriz de correlação


```

corr = dados_saude.corr()

print(corr)
```


# Interpretação

Podemos encontrar forte correlação entre:

- Exercício
- Perda de peso

Mas:
- Parte do efeito vem da alimentação.

Isso mostra como variáveis escondidas podem alterar conclusões.



# EXEMPLO 4 — Simpson’s Paradox

## Cenário

Às vezes a análise geral mostra um resultado,
mas quando dividimos os grupos o comportamento muda.


```

grupo_a_estudo = np.random.normal(3, 1, 50)
grupo_a_nota = grupo_a_estudo * 1.2 + 5

grupo_b_estudo = np.random.normal(8, 1, 50)
grupo_b_nota = grupo_b_estudo * 1.2 - 5

grupo = ['A'] * 50 + ['B'] * 50

horas = np.concatenate([grupo_a_estudo, grupo_b_estudo])
notas = np.concatenate([grupo_a_nota, grupo_b_nota])

dados_simpson = pd.DataFrame({
    'Grupo': grupo,
    'Horas': horas,
    'Nota': notas
})

dados_simpson.head()
```


# Visualização por grupo


```

plt.figure(figsize=(8,5))

for g in dados_simpson['Grupo'].unique():
    subset = dados_simpson[dados_simpson['Grupo'] == g]
    plt.scatter(subset['Horas'], subset['Nota'], label=g)

plt.xlabel('Horas')
plt.ylabel('Nota')

plt.title("Paradoxo de Simpson")

plt.legend()

plt.show()
```


# Conclusões importantes

## Em Ciência de Dados:

- Correlação NÃO prova causalidade
- Variáveis confundidoras podem enganar modelos
- Contexto é extremamente importante
- Modelos podem aprender padrões errados
- Interpretabilidade é fundamental
