# Aula 01 — Guia de Navegação

## Objetivo da aula

Demonstrar que o mesmo problema de negócio — recomendação de filmes — pode ser formulado como três paradigmas distintos de aprendizado de máquina:

- **Supervisionado:** aprender com rótulos explícitos.
- **Não supervisionado:** descobrir estrutura sem alvo definido.
- **Aprendizado por Reforço:** aprender por interação e recompensa.

## Narrativa unificadora — MovieLens

A turma assume o papel de equipe de dados de uma plataforma de streaming. O dataset MovieLens fornece:

- usuários com histórico de avaliações,
- filmes com metadados de gênero,
- notas explícitas como proxy de preferência.

A mesma base de dados gera três perguntas de negócio distintas, cada uma respondida por um paradigma diferente de ML.

**Mensagem central da aula:** o dataset é o mesmo. O que muda é a pergunta formulada ao sistema.

## Quadro comparativo dos três paradigmas

| Paradigma | Pergunta | Dado | Sinal | Saída |
|---|---|---|---|---|
| Supervisionado | O usuário vai gostar deste filme? | Pares com resposta conhecida | Rótulo | Predição de nota ou preferência |
| Não supervisionado | Que perfis de usuários ou filmes existem? | Observações sem alvo explícito | Estrutura interna | Clusters, segmentos, representações |
| Reinforcement Learning | Qual filme devo recomendar agora? | Interação + feedback | Recompensa | Política de recomendação (MDP) |

Fonte: Roteiro da Aula 01, seção 10.

## Conexão com o Desafio Final de Recomendação

O problema de recomendação MovieLens é o fio condutor de todo o curso. Esta aula o formula em três perspectivas complementares:

- **Supervisionado (NB02):** "O usuário vai gostar deste filme?" → preditor de preferência — o *ranqueador* de candidatos.
- **Não supervisionado (NB03):** "Que perfis de usuários existem?" → segmentação — comprime o espaço de estados para o agente de RL.
- **RL Bandit (NB04):** "Qual filme recomendar agora?" → política de decisão sequencial — o *decisor* que aprende com feedback.

No Desafio Final, os três componentes se combinam: o não supervisionado define estados, o supervisionado pré-filtra ações, e o agente de RL toma a decisão final e aprende com a interação ao longo do tempo.
